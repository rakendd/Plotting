# Temperature Gradient Matrix 
## with annotation on the plot 

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import Image
```

This is part of a project I was working on. <br>
For this I need to the temperature field on the ship helo-deck region for many cases and to ensure, 
the temperature over there has not exceeded the acceptable levels. 

There is a research paper for this with a plot showing Temperature Gradient Matrix. <br>

```python
Image(filename='tgm.png')
```

Now I have to digitize this plot and put which cases lie at these levels. <br>
This is a simple task, 
can do with a heat map filling the cell grid with the required color. <br>
We have three colors here, red, green and amber. <br>
In Seaborn, we can customize the color palette. 
We have three colors here, red, green and amber. <br>
These colors hex value can be found by uploading the image at https://imagecolorpicker.com/ and picking the color. <br>


### Customized Color palette

```python
Gr_Ye_Rd = ["#00AF50", "#FFC000","#FE0000"]
sns.palplot(sns.color_palette(Gr_Ye_Rd))
plt.show()
```

Make a zero array of size 8 x 5 <br>
and give the array values as 0 for green, 0.5 for amber and 1 for red

### Array for plot values

```python
data = np.zeros((8,5))
```

```python
data_mod = data.copy()

data_mod[2:,1:] = 0.5

data_mod[5:,3:] = 1

data_mod[3:5,4]= 1
```

```python
# Now the modified array looks like
data_mod
```

### For annotation on the plot

We annotate this plot with text, to say which particular case is falling on which grid cell of this matrix plot.<br>
For this, we will make an array of strings. 

```python
labels = np.array([['' for i in range(5)] for j in range(8)],dtype='U15')
```

```python
labels[5,1] = 'cases_1'
labels[6,1] = 'cases_2'
labels[5,2] = 'cases_3'
labels[6,2] = 'cases_4'
```

```python
labels
```

# Plotting

```python
# This is size of the figure
from matplotlib import rcParams
rcParams['figure.figsize'] = 10,5 
```

```python
import numpy as np
import seaborn as sns; sns.set()
ax = sns.heatmap(data_mod, cmap=sns.color_palette(Gr_Ye_Rd),linewidths=2, linecolor='white',cbar = False, annot=labels,fmt='',annot_kws={"size": 16})
plt.yticks(rotation=0)


# These are the labels we want to use instead of automated numerics 

ax.set_xticklabels(['0<=2','>2','>10','>30','>40'], fontsize=12)
ax.set_yticklabels(['<50', '<40', '<30', '<25', '<20', '<15', '<10', '<5'], fontsize=12)


# matplotlib.rcParams.update({'font.size': 22})
plt.xlabel('Temperature rise above ambient',fontsize=17)
plt.ylabel('Height above helo-deck (m)',fontsize=17)
plt.title('Tempearture Gradient Matrix',fontsize=20)
# plt.savefig('temp_matrix.png')
cbar_ax=None
plt.show()
```
